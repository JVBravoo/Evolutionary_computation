## Plot 

```python
import matplotlib.pyplot as plt
def plota(variacao, logbook):
    gen = logbook.select("gen")
    fit_min = logbook.select("min")
    fit_avg = logbook.select("avg")
    fit_max = logbook.select("max")

    fig, ax1 = plt.subplots()
    line1 = ax1.plot(gen, fit_min, "b", label="Minimum")
    ax1.set_xlabel("Generation")
    ax1.set_ylabel("Fitness", color="b")

    ax2 = ax1.twinx()
    line2 = ax2.plot(gen, fit_avg, "r", label="Average")
    ax2.set_ylabel("Size", color="r")

    lns = line1 + line2
    labs = ['Minimum', 'Average', 'Maximum']
    ax1.legend(lns, labs)

    plt.savefig("variacao_" + str(variacao) + ".png")
    plt.show()
```

```python
# toolbox.register("evaluate", evalOneMax)
# toolbox.register("mate", tools.cxTwoPoint)
# toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
# toolbox.register("select", tools.selTournament, tournsize=3)

# CXPB  is the probability with which two individuals are crossed
# indpb is the probability for each attribute to be exchanged
# MUTPB is the probability for mutating an individual

var_1 = {"population": 300, "ind": 100, "indpb": 0.05, "torneio": 3, "cxpb": 0.5, "mutpb": 0.2, "ngen": 100}
var_2 = {"population": 800, "ind": 100, "indpb": 0.05, "torneio": 3, "cxpb": 0.5, "mutpb": 0.2, "ngen": 100}
var_3 = {"population": 300, "ind": 50, "indpb": 0.05, "torneio": 3, "cxpb": 0.5, "mutpb": 0.2, "ngen": 100}
var_4 = {"population": 300, "ind": 100, "indpb": 0.05, "torneio": 7, "cxpb": 0.5, "mutpb": 0.2, "ngen": 100}
var_5 = {"population": 300, "ind": 100, "indpb": 0.05, "torneio": 3, "cxpb": 0.2, "mutpb": 0.2, "ngen": 100}
var_6 = {"population": 300, "ind": 100, "indpb": 0.05, "torneio": 3, "cxpb": 0.5, "mutpb": 0.6, "ngen": 100}
var_7 = {"population": 300, "ind": 100, "indpb": 0.05, "torneio": 3, "cxpb": 0.5, "mutpb": 0.2, "ngen": 50}
var_8 = {"population": 300, "ind": 100, "indpb": 0.05, "torneio": 3, "cxpb": 0.8, "mutpb": 0.2, "ngen": 100}
var_9 = {"population": 300, "ind": 100, "indpb": 0.05, "torneio": 3, "cxpb": 0.5, "mutpb": 0.1, "ngen": 100}
var_10 = {"population": 300, "ind": 100, "indpb": 0.02, "torneio": 3, "cxpb": 0.5, "mutpb": 0.2, "ngen": 100}

variacoes = [var_1, var_2, var_3, var_4, var_5, var_6, var_7, var_8, var_9, var_10]
```

```python
for i in range(len(variacoes)):
    var = variacoes[i]
    creator.create("FitnessMax", base.Fitness, weights=(1.0,))
    creator.create("Individual", list, fitness=creator.FitnessMax)

    toolbox = base.Toolbox()
    # Attribute generator 
    toolbox.register("attr_bool", random.randint, 0, 1)
    # Structure initializers
    toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, n=var["ind"])
    toolbox.register("population", tools.initRepeat, list, toolbox.individual)
    toolbox.register("evaluate", evalOneMax)
    toolbox.register("mate", tools.cxTwoPoint)
    toolbox.register("mutate", tools.mutFlipBit, indpb=var["indpb"])
    toolbox.register("select", tools.selTournament, tournsize=var["torneio"])

    stats = tools.Statistics(key=lambda ind: ind.fitness.values)
    stats.register("avg", np.mean)
    stats.register("min", np.min)
    stats.register("max", np.max)

    pop, logbook = algorithms.eaSimple(toolbox.population(n=var["population"]), toolbox, cxpb=var["cxpb"], mutpb=var["mutpb"], 
                                        ngen=var["ngen"], stats=stats, verbose=True)
    
    plota(i, logbook)
```
